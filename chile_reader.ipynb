```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
error = .15 # probably smaller than 0.15 K
```

## Longitude 0.5 kpc

```python
data5 = np.loadtxt('survey_files/Low-L spectra/L3.5_R0.5.txt')
plt.axhline(0, color='orange')
plt.plot(data5[:,0], data5[:,1])
```

```python
plt.axhline(0, color='orange')
plt.errorbar(data5[:,0], data5[:,1], error, fmt='.')
plt.xlim(260,330)
plt.ylim(-1,1)
```

```python
plt.axhline(0, color='orange')
plt.errorbar(data5[:,0], data5[:,1], error, fmt='.')
plt.xlim(280,285)
plt.ylim(-.25,.25)
plt.gca().set_xticks(np.arange(280,285,.1), minor=True)
plt.gca().grid(which='major', alpha=0.8)
plt.gca().grid(which='minor', alpha=0.3)
```
